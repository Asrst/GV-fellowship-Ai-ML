```python
print(test.shape)
print(temp.shape)
```

```python
from scipy.sparse import csr_matrix
temp_values = csr_matrix(temp.values)
```

```python
from sklearn.neighbors import NearestNeighbors
modelKnn =  NearestNeighbors(metric = 'cosine', algorithm = 'brute')
modelKnn.fit(temp_values)
```

```python
query_index = np.random.choice(temp.shape[0])
distances, indices = modelKnn.kneighbors(temp.iloc[query_index, :].values.reshape(1,-1), n_neighbors = 3)

for i in range(0, len(distances.flatten())):
    if i==0:
        print('users with simialr taste as userId={0}:\n'.format(temp.index[query_index]))
    else:
        a = np.array(temp.iloc[query_index].values)
        b = np.array(temp.iloc[indices.flatten()[i]].values)
        c = np.array(list(map(operator.sub, a, b)))
        d = np.where(c > 0)
        print('{0}: for userId:{1}, I recommend movieIds:{2} as distance:{3}'.format(i, temp.index[indices.flatten()[i]], d, 
                                                                    distances.flatten()[i]))
```

```python
#query_index = np.random.choice(temp.shape[0])

res = pd.DataFrame(columns=temp.columns, index = temp.index)

for query_index in range(0, len(temp)):
    distances, indices = modelKnn.kneighbors(temp.iloc[query_index, :].values.reshape(1,-1), n_neighbors = 2)
    for i in range(0, len(distances.flatten())):
        if i==0:
            print('users with simialr taste as userId={0} is:'.format(temp.index[query_index]))
        else:
            res.iloc[query_index] = temp.iloc[query_index] - temp.iloc[indices.flatten()[i]]
            print('userId:{0} at distance:{1} \n'.format(temp.index[indices.flatten()[i]],
                                                                    distances.flatten()[i]))
```

#### Question-1: 
#### Write a function to compute euclidean distance between two users for all entries except the missing yth entry. 
#### We will use KNN tofind the K nearest neighbour users.

```python
from scipy.spatial import distance 
from sklearn.metrics.pairwise import pairwise_distances

def getSimilarUsers(user,num,metric):
    temp = pd.pivot_table(data = itrain, index='userId', values='rating', columns='movieId',aggfunc='sum', fill_value=0)
    df_dist=pd.DataFrame(data = pairwise_distances(temp.values,metric=metric))
    print("\n\nUserId \t Distance ")
    return df_dist[user].sort_values().head(num)

user=int(input("Enter User Id : "))
num=int(input("Enter Number of Similar Users : "))
print(getSimilarUsers(user,num,'euclidean'))
```

```python
from scipy.spatial import distance 
from sklearn.metrics.pairwise import pairwise_distances

def getSimilarUsers(user,num,metric):
    temp = pd.pivot_table(data = itrain, index='userId', values='rating', columns='movieId',aggfunc='sum', fill_value=0)
    df_dist=pd.DataFrame(data = pairwise_distances(temp.values,metric=metric))
    print("\n\nUserId \t Distance ")
    return df_dist[user].sort_values().head(num)

user=int(input("Enter User Id : "))
num=int(input("Enter Number of Similar Users : "))
print(getSimilarUsers(user,num,'cityblock'))
```
