### Exploring the Linrigress data set: It has two coloumns (X and Y) with 90 rows 

```python
#import libraries
import os
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import xlrd
from scipy import stats
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("/home/sa05975666/07-08-2018/Day_02_L4.xlsx", header = None)
```

```python
df1.rename(columns = {0:'X', 1:'Y'}, inplace = True)
```

```python
df1['Y_sq'] = df1.Y ** 2
```

### a) Load and plot values on a graph. Plot it as connected dots(line) and scattered dots.

```python
#sns.regplot(x="X", y="Y", data = df1)
plt.plot(df1['X'], df1['Y'], 'o', label='data points')
plt.plot(df1['X'], df1['Y'], label = 'line')
```

### b) Plot x vs y^2 graph

```python
df1[['X','Y_sq']].set_index('Y_sq').plot.line()
```

```python
plt.plot(df1['X'], df1['Y_sq'])
```

### c) Plot best fit line using stats library.
### d) Compare the best fit line with another line joining the two extremes, any two points.

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df1['X'], df1['Y'])
```

```python
plt.plot(df1['X'], df1['Y'], 'o', label='data points')
plt.plot(df1['X'], intercept + slope*df1['X'], 'g', label='best fit line')
print(slope, intercept, r_value, p_value, std_err)

ext = df1[['X', 'Y']][::88]
plt.plot(ext['X'], ext['Y'], label = 'line of extremes')
plt.legend()
```

### e) Change value of y and see the effect of outliers on the line.


```python
df1['Y'][43, 70] = 3.21
df1['Y'][25, 52] = -1.23

slope, intercept, r_value, p_value, std_err = stats.linregress(df1['X'], df1['Y'])
plt.plot(df1['X'], df1['Y'], 'o', label='data points')
plt.plot(df1['X'], intercept + slope*df1['X'], 'g', label='Effect of outliers')

print(slope, intercept, r_value, p_value, std_err)
```

### f) Store every 10th value and find best fit curve.

```python
tv = df1[::10]
slope, intercept, r_value, p_value, std_err = stats.linregress(tv['X'], tv['Y'])
plt.plot(tv['X'], tv['Y'], 'o', label='data points')
plt.plot(tv['X'], intercept + slope*tv['X'], 'g', label='reg line for 10th values')

print(slope, intercept, r_value, p_value, std_err)
```
