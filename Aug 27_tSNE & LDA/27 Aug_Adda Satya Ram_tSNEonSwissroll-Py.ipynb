```python
from sklearn.datasets import make_swiss_roll
train = make_swiss_roll(n_samples = 15000)

X = pd.DataFrame(data = train[0])
print(X.shape)
y = train[1]
len(y)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize = (6,4))
ax = plt.axes(projection='3d')
ax.scatter3D(X[0],X[1],X[2], c = y)
ax.set_xlabel('X[0]')
ax.set_ylabel('X[1]')
ax.set_zlabel('X[2]')
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2, verbose=1, perplexity = 5, n_iter = 500)
tsne_results = tsne.fit_transform(X)

df_tsne = X.copy()
df_tsne['y'] = y
df_tsne['x-tsne'] = tsne_results[:,0]
df_tsne['y-tsne'] = tsne_results[:,1]

plt.scatter(df_tsne['x-tsne'],df_tsne['y-tsne'],c = df_tsne['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('perplexity = 30, n_iter = 1000')
```

### Excersice 2: With noise in Swill Roll Data

```python
# loading swiss data
from sklearn.datasets import make_swiss_roll

X1 , Y1 = make_swiss_roll(n_samples=1500, noise=0.20, random_state=124)
X1 = pd.DataFrame(X1)
Y1 = pd.DataFrame(Y1)
```
