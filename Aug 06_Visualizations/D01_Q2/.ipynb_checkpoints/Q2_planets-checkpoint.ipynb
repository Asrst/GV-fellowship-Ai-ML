```python
#import libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import xlrd
import matplotlib.pyplot as plt
```

```python
planets_df = sns.load_dataset("planets")
planets_df.head(10)
```

### Which year has highest no.of measures?

```python
bp = sns.factorplot("year", data=planets_df, aspect=1.5,
                       kind="count", color='orange')
bp.set_xticklabels(step=3)
```

### Year wise & planet wise count of all 7 planets

```python
bp = sns.factorplot("year", data=planets_df, aspect=3,
                    hue='number', kind="count", order=range(2001, 2015))
bp.set_xticklabels(step=1)
```

### How mass of planets related with distance it travelled

```python
sns.jointplot(x="mass", y="distance", data=planets_df)
```

```python
sns.jointplot(x="number", y="orbital_period", data=planets_df)
```

### PairGrid to get glance at Data

```python
g = sns.PairGrid(planets_df, vars=["mass", "distance", 'orbital_period', 'year'], hue="number")
g.map(plt.scatter)
g.add_legend()
```
