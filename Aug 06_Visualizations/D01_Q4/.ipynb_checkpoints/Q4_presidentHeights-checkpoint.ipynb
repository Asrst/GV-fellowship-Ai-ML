### Data Visualization: Height of US Presidents Data

```python
#import libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import xlrd
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
df = pd.read_excel('/home/sa05975666/06-08-2018/D01_Q4/Q4_president_Heights.xlsx')
```

### Which number president is of highest and least height ? 
### Highest is 16th and 36th president
### Least is 4th president

```python
plt.figure(figsize=(30,10))
plt.xticks(df['order'])
ax =plt.plot('order', 'height(cm)', data = df)
```

```python
df[['order','height(cm)']].groupby(['height(cm)'], as_index = True)[['population']].sum().plot.bar(figsize=(15,5))
```

```python
df_sorted = df.sort_values(by = 'height(cm)', ascending = True)
df_sorted['order'] = df_sorted['order'].astype(str)
```

### Who is of great height and least height
### Highest is Lincoln and Lyndon B.Jonhson
### Least is James Madinson

```python
df_sorted.set_index('name').plot.bar(figsize = (15,5))
```
