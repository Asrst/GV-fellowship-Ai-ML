### Data Visualization: Seattle data (time series)

```python
#import libraries
import os
import pandas as pd
import numpy as np
import seaborn as sns
import xlrd
import matplotlib.pyplot as plt
```

```python
#get & set working directory
os.getcwd()
```

```python
# read the data files
df = pd.read_excel("/home/sa05975666/06-08-2018/D01_Q3/Q3_Seattle.xlsx")
```

```python
# find out what's there in data set
print(df.shape)
print(df.columns)
#df.describe()
```

```python
display(df.head())
print(df.STATION.unique())   # redundant as only of single type - no variance
df.STATION_NAME.unique() ## redundant as only of single type - no variance
```

```python
df = df.drop(['STATION', 'STATION_NAME'], axis = 1)
```

```python
# changing the date coloum to proper date format using pandas date time
df['dd'] = pd.to_datetime(df['DATE'], format='%Y%m%d', errors='ignore')
df = df.drop(['DATE'], axis = 1)
```

### PRCP is high for a day in march 
## There are also few other days with peaks 

```python
#time-series : PRCP v/s DATE
df[['PRCP', 'dd']].set_index('dd').plot(figsize=(10,5), grid=True)
```

```python
#time-series : SNOW v/s DATE
df[['SNWD', 'SNOW', 'dd']].set_index('dd').plot(figsize=(10,5), grid=True)
```
