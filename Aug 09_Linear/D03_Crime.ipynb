## Crime Rate Data Set

```python
#import libraries
import os
import pandas as pd
import numpy as np
import scipy as sp
import seaborn as sns
import xlrd
from scipy import stats
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
# set work directory
path = os.getcwd()
```

```python
#read data
df = pd.read_excel(str(path) + "/D03_Crime_data.xlsx")
```

```python
df.isnull().any()
```

### Running Blank model with all variables
### This set our base value of metrics so that we can improve them

```python
target = df['X1']
predictors = df.drop('X1', axis = 1)

# Test, train splits
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(predictors, target, test_size=0.2, random_state=7)

#X_train, X_test = predictors[:-12], predictors[-12:]
#Y_train, Y_test = target[:-12], target[-12:]
```

```python
# dummy model
from sklearn import linear_model
from sklearn import metrics

# Creating a linear regression object
model = linear_model.LinearRegression()

#fitting the model
model.fit(X_train, Y_train)

#predicting 
Y_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
print('Values of Coefficients:', model.coef_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(Y_test, Y_pred))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(Y_test, Y_pred))

#mae
print('mae:', mean_absolute_error(Y_test, Y_pred))
```

### modeling with new preditors 
### Removing the X2: Voilent Crime Rate - as it is dependent variable
### result: reduce in the Error

```python
predictors1 = df[['X3','X4', 'X5', 'X6','X7']]

# model
from sklearn import linear_model
from sklearn import metrics

# Test, train splits
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(predictors1, target, test_size=0.2, random_state = 7)

# Creating a linear regression object
model = linear_model.LinearRegression()
#fitting the model
model.fit(X_train, Y_train)
#co-efficients
print('Values of Coefficients:', model.coef_)

#prediction
y_train_pred = model.predict(X_train)
Y_pred = model.predict(X_test)


from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(Y_test, Y_pred))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(Y_test, Y_pred))

#mae
print('mae:', mean_absolute_error(Y_test, Y_pred))
```
