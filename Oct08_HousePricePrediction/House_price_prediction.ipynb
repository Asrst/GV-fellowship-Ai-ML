```python
def analyze_for_miss_val(data):
    
    missing_val_col = []
    for col in data.columns:
        if data[col].isnull().sum() != 0:
            missing_val_col.append(col)
            
    
    miss_val = data[missing_val_col].isnull().sum()

#     col_tobe_removed = []
#     for row in miss_val.index:
#         if (miss_val[row]/data.shape[0] * 100) > 90:
#             col_tobe_removed.append(row)
    
    return miss_val
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, r2_score


def get_score(predicted, original):    
    r2 = r2_score(predicted, original)
    rmse = np.sqrt(mean_squared_error(predicted, original)) 
    return r2, rmse


def build_model(model, x_trn, x_tst, y_trn, y_tst):
    
    model = model.fit(x_trn, y_trn)
    
    #train-prediction & results
    y_trn_pred = model.predict(x_trn)
    trn_r2, trn_rmse = get_score(y_trn_pred, y_trn)
    
    # results on validation set
    y_pred = model.predict(x_tst)
    test_r2, test_rmse = get_score(y_pred, y_tst)
    
    return trn_r2, trn_rmse, test_r2, test_rmse

def feature_imp(model):
    features = predictors.columns
    
    importances = model.feature_importances_
    indices = np.argsort(importances[:20])
    
    fImp_Asc = []
    
    for i in indices:
        n = features[i]
        fImp_Asc.append(n)
        
    plt.figure(figsize=(5,5))
    plt.barh(range(len(indices)), importances[indices], color='b', align='center')
    plt.yticks(range(len(indices)), [features[i] for i in indices])
    plt.xlabel('Relative Importance')
    plt.title('Feature Importances')
    plt.show()
    

def model_with_kfold(model, X, y, folds):
    
    print('Model Summary: \n{}' .format(model))
    
    print('\n Data Folds: {}'.format(folds))
    
    train_R2 = []
    train_RMSE = []
    test_R2 = []
    test_RMSE = []
    
    kf = KFold(n_splits = folds, random_state = 123) 
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        trn_r2, trn_rmse, test_r2, test_rmse = build_model(model, X_train, X_test, y_train, y_test)
        
        train_R2.append(trn_r2)
        train_RMSE.append(trn_rmse)
        test_R2.append(test_r2)
        test_RMSE.append(test_rmse)

    print('\n TRAINING:')   
    print('KFold R2: {}' .format(np.array(train_R2).mean()))
    print('KFold RMSE: {}' .format(np.array(train_RMSE).mean()))
    
    print('\n VALIDATION: ')
    print('KFold R2: {}' .format(np.array(test_R2).mean()))
    print('KFold RMSE: {}' .format(np.array(test_RMSE).mean()))
    
    #feature_imp(model)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler

rc = RobustScaler(copy=True, quantile_range=(5.0, 95.0), with_centering=True, with_scaling=True)
X = rc.fit_transform(train_data)

#X = np.array(train_data)

y = np.array(train_labels)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression(n_jobs = -1)
model_with_kfold(lr, X, y, 5)
```
