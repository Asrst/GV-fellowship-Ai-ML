```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.utils import to_categorical
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)

# flatten 28*28 images to a 784 vector for each image
num_pixels = X_train.shape[1] * X_train.shape[2]
X_train_flattened = X_train.reshape(X_train.shape[0], num_pixels).astype('float32')
X_test_flattened = X_test.reshape(X_test.shape[0], num_pixels).astype('float32')
```

```python
num_pixels
```

```python
# normalize inputs from 0-255 to 0-1
X_train_std = X_train_flattened / 255
X_test_std = X_test_flattened / 255

# one hot encode outputs
ytrain_cat = to_categorical(y_train)
ytest_cat = to_categorical(y_test)
num_classes = ytrain_cat.shape[1]
```

```python
print(ytrain_cat[0])
print(ytrain_cat[1])
print(ytrain_cat[2])
```

```python
#### TSNE

from sklearn.manifold import TSNE

tsne = TSNE(n_components = 2, verbose=1, perplexity = 40, n_iter = 1000, random_state = 7, learning_rate = 1000)
tsne_results = tsne.fit_transform(X_train_std[:10000])

df_tsne = pd.DataFrame()
df_tsne['y'] = y_train[:10000]
df_tsne['x-tsne'] = tsne_results[:,0]
df_tsne['y-tsne'] = tsne_results[:,1]
```

```python
sns.lmplot(x = 'x-tsne', y= 'y-tsne', data = df_tsne, hue = 'y', 
           legend = True, fit_reg = False)
#scatter_kws = {'s': 200, 'alpha':0.3}

plt.title('perplexity = 100, n_iter = 1000')
plt.xlabel('x-tsne -dim 1')
plt.ylabel('y-tsne -dim2')
plt.show()
```

### Base model : Using 3 layers: relu, tanh and softmax

```python
# define baseline model
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(num_pixels, input_dim = num_pixels, kernel_initializer='normal', activation='relu'))
    model.add(Dense(500, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

# build the model
base_ann = baseline_model()
# Fit the model
base_ann = base_ann.fit(X_train_std, ytrain_cat, validation_data=(X_test_std, ytest_cat), epochs=10, batch_size = 200, 
                        verbose=1)

# # Final evaluation of the model
# scores = base_ann.evaluate(X_test_std, ytest_cat, verbose=0)
# print("Baseline Error: %.2f%%" % (100-scores[1]*100))

## Get the accuarcies
train_acc = base_ann.history['acc']
validation_acc = base_ann.history['val_acc']

## plot the accuaracies
plt.figure(figsize = (6,4))
plt.plot(np.arange(1,11,1), train_acc, 'o-', label='train Acc')
plt.plot(np.arange(1,11,1), validation_acc, 'o-', label='valid Acc')
plt.xticks(np.arange(1,11,1))
plt.xlabel('no.of epoch')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.legend()
plt.show()
```

### Build_model

```python
def build_model(model_name):
    # build the model
    ann_model = model_name
    # Fit the model
    ann_model = ann_model.fit(X_train_std, ytrain_cat, validation_data=(X_test_std, ytest_cat), epochs=10, 
                              batch_size = 200, verbose=1)

    train_acc = ann_model.history['acc']
    validation_acc = ann_model.history['val_acc']

    plt.figure(figsize = (6,4))
    plt.plot(np.arange(1,11,1), train_acc, 'o-', label='train Acc')
    plt.plot(np.arange(1,11,1), validation_acc, 'o-', label='valid Acc')
    plt.xticks(np.arange(1,11,1))
    plt.xlabel('no.of epoch')
    plt.ylabel('Accuracy')

    plt.tight_layout()
    plt.legend()
    plt.show()
```

### First model: Using relu as activation

```python
# define baseline model
def first_model():
    # create model
    model = Sequential()
    model.add(Dense(num_pixels, input_dim = num_pixels, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1000, kernel_initializer='normal', activation='relu'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

build_model(first_model())
```

### Second model using : tanh as activation

```python
# define baseline model
def second_model():
    # create model
    model = Sequential()
    model.add(Dense(num_pixels, input_dim = num_pixels, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(500, kernel_initializer='normal', activation='tanh'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

build_model(second_model())
```

### Third model - using sigmoid as activation

```python
# define baseline model
def third_model():
    # create model
    model = Sequential()
    model.add(Dense(num_pixels, input_dim = num_pixels, kernel_initializer='normal', activation='hard_sigmoid'))
    model.add(Dense(500, kernel_initializer='normal', activation='hard_sigmoid'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

build_model(second_model())
```

### Neural layer Tuning

```python
# define baseline model - sgd 
def fourth_model():
    # create model
    model = Sequential()
    model.add(Dense(num_pixels, input_dim = num_pixels, kernel_initializer='normal', activation='relu'))
    model.add(Dense(num_classes, kernel_initializer='normal', activation='softmax'))
    # Compile model
    model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
    return model

build_model(fourth_model())
```
